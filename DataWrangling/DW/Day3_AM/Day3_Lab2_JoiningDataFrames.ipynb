```python
df_flights = pd.DataFrame(dict(
    AIRPORT_CODE=['MIA','MIA','LAX','DCA'],
    FLIGHT_NO=['3322','3213','4223','5678']
))
df_airports = pd.DataFrame(dict(
    AIRPORT_CODE=['MIA','LAX','DCA','HEL','SFO'],
    AIRPORT_NAME=['Miami International Airport', 
                  'Los Angeles International Airport', 
                  'Ronald Reagan Washington',
                  'Helsinki-Vantaa Airport',
                  'San Francisco International Airport']))
df_flights
```

```python
df_airports
```

```python
#merge using common column
df_flights.merge(df_airports, left_on = 'AIRPORT_CODE', right_on = 'AIRPORT_CODE')
```

```python
#directly using pd 
pd.merge(df_flights, df_airports , on = 'AIRPORT_CODE' )
```

```python
pd.merge(df_flights, df_airports , on = 'AIRPORT_CODE', how='inner' )
```

```python
pd.merge(df_flights, df_airports , on = 'AIRPORT_CODE', how='left' ) # right, outer
```

```python
pd.merge(df_flights, df_airports , on = 'AIRPORT_CODE', how='outer' )
```

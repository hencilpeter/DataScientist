```python
df = pd.read_csv('stock.csv')
df
```

```python
df['rating'] = df['Stock Price'] 
df['Event'] = np.where((df.Event == 'Painting'),'Art',df.Event)
```

```python
df['rating'] = np.where((df['Stock Price']  < 200), 'Low', 
                        np.where(( (df['Stock Price']  > 200) & (df['Stock Price']  < 400) ), 'Normal', 'High'))  
df
```

```python
df['rating'] = np.where((df['Stock Price']  < 200), 'Low', 
                        np.where((df['Stock Price']  < 400 ), 'Normal', 'High'))  
df
```

```python
#another approach 
def rate(price):
    if price <200:
        return 'Low'
    if price <400:
        return 'Normal'
    
    return 'High'

df['Stock Price'].map(rate)
```

```python
df['Stock Price'].map(lambda price: 'Low' if price <200 else 'Normal' if price < 400 else 'High')
```

```python
def getFirstAndLastName(name):
    nameparts = name.split(" ") 
    return f'{nameparts[0]} { "-".join(nameparts[1:])}' 

df['name'] =df['name'].map(getFirstAndLastName)
df['name'].str.split(expand=True)
```

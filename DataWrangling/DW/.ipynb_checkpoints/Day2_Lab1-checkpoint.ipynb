```python
 #id_vars - which column you want to keep, 
#var_name = name ofthe new col
#value_vars - what value to be put in 
df_melted = df.melt(id_vars =['Country','City'],
                   var_name = 'Date',
                   value_vars =  df.columns[2:],
                   value_name= 'Temp')
```

```python
df_melted = df.melt(id_vars =['Country','City'])
df_melted
```

```python
df_melted.columns = ['Country','City','Date', 'Temp']
df_melted
```

```python
#to printout the temperature for each country
df_melted
```

```python
groups
```

```python
groups = df_melted.groupby("Country")
for country,group in groups:
    print(country)
    print('-' * len(country))
    for city, citygroup in group.groupby("City"):
        print('    ' + city)
        print(citygroup['Temp'].sort_values().to_string(index=False))
```

```python
for (country_city, gp) in df_melted.groupby(['Country','City']):
    print(country_city)
    print(gp[['Date','Temp']])
```

```python
groups = df_melted.groupby("Country")
for country,group in groups:
    print(country)
    print('-' * len(country))
    for city, citygroup in group.groupby("City"):
        print('    ' + city)
        print('    ' + '=' * len(city))
        print(citygroup['Temp'].mean())
```

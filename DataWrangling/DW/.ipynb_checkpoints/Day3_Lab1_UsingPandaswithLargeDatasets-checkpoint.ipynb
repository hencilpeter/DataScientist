```python
import pandas as pd
```

```python
#file size is too huge but read only 900
df_chunk = pd.read_csv('df.csv', chunksize=900)

def process_chunk(df):
    #filter out the column and retain the columns "2" and "4"
    #return df[["2","4"]]
    return df.iloc[:][['2','4']]
    #return df.iloc[::2][['2','4']]  #only alternative rows' columns
    
#every chunk will be read and passed to the functions     
for chunk in df_chunk:
    print(process_chunk(chunk))
```

```python
#join all the DF into single data frame 
df_chunk = pd.read_csv('df.csv', chunksize=900)

def process_chunk(df):
    #filter out the column and retain the columns "2" and "4"
    #return df[["2","4"]]
    return df.iloc[:][['2','4']]
    #return df.iloc[::2][['2','4']]  #only alternative rows' columns
    
#every chunk will be read and passed to the functions  
resultDF = pd.DataFrame()
for chunk in df_chunk:
    resultDF = resultDF.append(process_chunk(chunk), sort=False)

print(resultDF)
```

```python
#another solution 
import pandas as pd
df_chunk = pd.read_csv('df.csv', chunksize=900)
df_chunk    
def process_chunk(df):
    return df.iloc[:][['2','4']]
    
chunk_list = []    
for chunk in df_chunk:
    chunk_list.append(process_chunk(chunk))
    
df_concat = pd.concat(chunk_list)
df_concat
```

```python
#3rd solution 
import pandas as pd
df_chunk = pd.read_csv('df.csv', chunksize=900)
df_chunk    
def process_chunk(df):
    return df.iloc[:][['2','4']]
    
result_DF = pd.DataFrame() 
for chunk in df_chunk:
    result_DF = pd.concat([result_DF,process_chunk(chunk)], axis=0)

result_DF
```

```python
import pandas as pd
```

```python
#only loading few columns while readign csv
df = pd.read_csv('airports.csv', usecols=['IATA_CODE', 'AIRPORT'])
df
```

```python
#test usecols with callable - 1st step 
def test(x):
    print(x)
    return x

df = pd.read_csv('airports.csv', usecols=test)
df
```

```python
#select only few columns
def test(x):
    if x == 'AIRPORT' or x == 'IATA_CODE':
        return x

df = pd.read_csv('airports.csv', usecols=test)
df
```

```python
#dont load lattitude and longitude
def test(x):
    if x != 'LATITUDE' and x != 'LONGITUDE':
        return x
df = pd.read_csv('airports.csv', usecols=test)
df
```

```python
def test(x):
    #if x == 'LATITUDE' or x == 'LONGITUDE':
    if x not in ('LATITUDE' , 'LONGITUDE'):
        return x
    

df = pd.read_csv('airports.csv', usecols=test)
df
```

```python
df = pd.read_csv('airports.csv', usecols= lambda x: x not in ('LATITUDE' , 'LONGITUDE') )
df
```

```python

LineCount=len(open('flights.csv').readlines()) 
df=pd.read_csv('flights.csv', skiprows=range(1,LineCount-20))
df
```

```python
#another way instead of using readlines 
#below one - loop takes time but it reads one by one. best suitable for huge file
LineCount = sum([1 for l in open('flights.csv')])
# df=pd.read_csv('flights.csv', skiprows=range(1,LineCount-20))
df=pd.read_csv('flights.csv', skiprows= lambda x : 0 < x < LineCount - 20 )
df
```

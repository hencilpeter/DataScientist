```python
import pandas as pd
from datetime import datetime
```

```python

#strptime() - convert a string into a datetime object
#strftime() - convert a datetime object into a string
#'2018-12-24' to 'Dec 24, 2018'
#'2018-12-24' --> date time object(strptime) --> 'Dec 24, 2018'(strftime)

df = pd.read_csv('AAPL.csv', 
                converters = {
                    'Close':lambda x: f'${float(x):.2f}',
                    'Date':lambda x: datetime.strptime(x,'%Y-%m-%d').strftime('(%A)%b %d, %Y')
                })
df
```
